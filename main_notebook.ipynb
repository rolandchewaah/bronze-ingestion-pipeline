```python
# Databricks notebook source
# MAGIC %md
# MAGIC # Ingestion Pipeline: Bronze Layer
# MAGIC **Source:** GitHub Repo `databricks-ingestion-platform`

# COMMAND ----------

# Setup Database and Table names
dbutils.widgets.text("env", "dev")
env = dbutils.widgets.get("env")

db_name = f"ingestion_platform_{env}"
table_name = "raw_traffic_data"

spark.sql(f"CREATE DATABASE IF NOT EXISTS {db_name}")
spark.sql(f"USE {db_name}")
```

```python
# MAGIC %md
# MAGIC ### 1. Sample Ingestion Logic (Auto Loader)
# MAGIC This block simulates ingesting JSON data from a landing zone.

# Sample data path (replace with your S3/ADLS mount point in production)
source_path = "/databricks-datasets/structured-streaming/events/"
checkpoint_path = f"/tmp/checkpoints/{env}/{table_name}"

# Auto Loader Ingestion
df = (spark.readStream
      .format("cloudFiles")
      .option("cloudFiles.format", "json")
      .option("cloudFiles.schemaLocation", f"{checkpoint_path}/schema")
      .load(source_path))
```

```python
# Write to Delta Table
(df.writeStream
 .format("delta")
 .option("checkpointLocation", f"{checkpoint_path}/data")
 .outputMode("append")
 .trigger(availableNow=True) # Runs once and stops, ideal for CI/CD
 .toTable(table_name))
```

```python
# MAGIC %md
# MAGIC ### 2. Verification
# MAGIC Show the last 10 rows ingested.

# COMMAND ----------

display(spark.sql(f"SELECT * FROM {table_name} ORDER BY timestamp DESC LIMIT 10"))
```
